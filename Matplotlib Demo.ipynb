```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5
y1 = 5*x + 6
#changing the figure
fig=plt.figure(figsize=(10,5))
#plotting the datapoints
plt.plot(x,y,linewidth = 2.0, linestyle = ":", color = 'r', alpha = 0.5, marker = 'o')
plt.plot(x,y1,linewidth = 2.0, linestyle = ":", color = 'blue', alpha = 0.8, marker = 'x')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1','myLine'],loc = 'best')
plt.grid(True)
plt.show()
```

```python
#importing the libraries
import matplotlib.pyplot as plt
%matplotlib inline

#creating the dataset
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

#creating the scatter plot
plt.scatter(a, b, color = 'g', s = 200, edgecolor = 'black', marker = 'o', alpha = 1)
plt.scatter(a, x, c = 'r', s = 400, edgecolor = 'b', marker = '4', alpha = 1)
plt.legend(['b','x'], loc = 'best')
plt.title('Scatter Plot Demo')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig('Scatter Plot.png')
plt.show()
```

```python
#importing the libraries
import matplotlib.pyplot as plt
%matplotlib inline

#data preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order, discount])
#plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot Demo")
plt.grid(True)
plt.show()
```

```python
#importing the libraries
import matplotlib.pyplot as plt
%matplotlib inline

#preparing the dataset
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

#creating the pie chart
plt.pie(sizes, labels = label, autopct = '%1.1f%%', shadow = True, startangle = 90, explode = (0.1,0.1,0,0))
plt.title("Pie Chart Demo")
plt.show()
```

```python
#importing the libraries
import matplotlib.pyplot as plt
%matplotlib inline

#define the dataset
group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_center = [5]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#creating the doughnut chart
pie1 = plt.pie(group_size,labels = group_names, radius = 1.5, colors = [a(0.5),b(1.5),c(1.0)])
pie2 = plt.pie(size_center,radius = 1.0,colors = 'w')
plt.show()
```
